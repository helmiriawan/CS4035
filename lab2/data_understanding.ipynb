```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Configuration
file_directory = "data_set" + "/"
training_file = file_directory + "swat_normal_dataset.csv"
testing_file = file_directory + "swat_attack_dataset.csv"
```

```python
# Load input file
training_data = pd.read_csv(training_file)
testing_data = pd.read_csv(testing_file)
```

```python
# Check data type of all columns
training_data.info()
```

```python
# Show sample data
training_data.head()
```

```python
# Generate descriptive statistics that summarize training data
training_data.describe()
```

```python
# Generate descriptive statistics that summarize testing data
testing_data.describe()
```

```python
# Change data type of a few features in training data
training_data['Timestamp'] = pd.to_datetime(training_data.Timestamp, errors='coerce')
training_data['Status'] = training_data['Status'].astype('category')

# Change data type of a few features in testing data
testing_data['Timestamp'] = pd.to_datetime(testing_data.Timestamp, errors='coerce')
testing_data['Status'] = testing_data['Status'].astype('category')
```

```python
# Visualize a few features of training data
training_data.plot(x='Timestamp', y='FIT101', figsize=(15, 4))
training_data.plot(x='Timestamp', y='LIT101', figsize=(15, 4))
training_data.plot(x='Timestamp', y='FIT201', figsize=(15, 4))
```

```python
# Visualize the first 10 hours of training data, when the system is just started
training_data[training_data['Timestamp'] < '2015-12-23 01:23:59'].plot(x='Timestamp', y='FIT101', figsize=(15, 4), color='orange')
training_data[training_data['Timestamp'] < '2015-12-23 01:23:59'].plot(x='Timestamp', y='LIT101', figsize=(15, 4), color='orange')
training_data[training_data['Timestamp'] < '2015-12-23 01:23:59'].plot(x='Timestamp', y='FIT201', figsize=(15, 4), color='orange')
```

```python
# Visualize a few features of training data in a day
training_data[(training_data['Timestamp'] > '2015-12-24 23:23:59') & (training_data['Timestamp'] < '2015-12-25 23:23:59')].plot(x='Timestamp', y='FIT101', figsize=(15, 4))
training_data[(training_data['Timestamp'] > '2015-12-24 23:23:59') & (training_data['Timestamp'] < '2015-12-25 23:23:59')].plot(x='Timestamp', y='LIT101', figsize=(15, 4))
training_data[(training_data['Timestamp'] > '2015-12-24 23:23:59') & (training_data['Timestamp'] < '2015-12-25 23:23:59')].plot(x='Timestamp', y='FIT201', figsize=(15, 4))
```

```python
# Visualize a few features of testing data in a day
testing_data[(testing_data['Timestamp'] > '2015-12-28 23:23:59') & (testing_data['Timestamp'] < '2015-12-29 23:23:59')].plot(x='Timestamp', y='FIT101', figsize=(15, 4), color='green')
testing_data[(testing_data['Timestamp'] > '2015-12-28 23:23:59') & (testing_data['Timestamp'] < '2015-12-29 23:23:59')].plot(x='Timestamp', y='LIT101', figsize=(15, 4), color='green')
testing_data[(testing_data['Timestamp'] > '2015-12-28 23:23:59') & (testing_data['Timestamp'] < '2015-12-29 23:23:59')].plot(x='Timestamp', y='FIT201', figsize=(15, 4), color='green')
```
